```python
df3.rename(columns = {'Primary Office':'Office'}, inplace = True)
df2.rename(columns = {'Primary Producer.1':'Primary Producer'}, inplace = True)
df3.rename(columns = {'Account Name.1':'Account Name'}, inplace = True)
```

```python
new_df = pd.merge(df1, df2, how= "inner", on = "Primary Producer")
```

```python
final_df = pd.merge(new_df, df3, how= "inner", on = "Account Name")
```

```python
final_df.drop(['Office_y'], axis=1, inplace =True)
```

```python
sns.boxplot(final_df["Annual Revenue"])
```

```python
sns.histplot(final_df["Annual Revenue"], kde = True)
```

```python
q3 = final_df["Annual Revenue"].quantile(0.75)
q1 = final_df["Annual Revenue"].quantile(0.25)
iqr = q3-q1
upper = q3 + (1.5*iqr)
lower = q1 - (1.5*iqr)

for i in final_df["Annual Revenue"]:
  if i > upper:
    final_df["Annual Revenue"] = final_df["Annual Revenue"].replace(i, final_df["Annual Revenue"].mean())
```
